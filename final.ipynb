```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def load_data(metadata_path, data_path):
    metadata = pd.read_csv(metadata_path)
    df = pd.read_csv(data_path)
    df = df.set_index('Sample Name')
    return metadata, df



def preprocess_data(df):
    concentration_cols = [col for col in df.columns if 'concentration_normalized' in col]
    if not concentration_cols:
        raise ValueError("No concentration_normalized columns found.")
    return df[concentration_cols]



def run_kmeans(df, k):
    X = df.values
    kmeans = KMeans(n_clusters=k, random_state=42, init="k-means++", n_init=10).fit(X)
    results_df = pd.DataFrame({'Sample Name': df.index, 'Cluster': kmeans.labels_})
    cluster_centers_df = pd.DataFrame(kmeans.cluster_centers_, columns=df.columns)

    feature_importance = cluster_centers_df.max() - cluster_centers_df.min()
    
    total_importance = feature_importance.sum()
    if total_importance > 0:
        feature_importance_standardized = feature_importance / total_importance
    else:
        feature_importance_standardized = feature_importance
        
    return results_df, cluster_centers_df, feature_importance_standardized



def analyze_data(metadata_path, data_path, k):
    metadata, df = load_data(metadata_path, data_path)
    numerical_df = preprocess_data(df)
    kmeans_results, cluster_centers, feature_importance_series = run_kmeans(numerical_df, k)
    merged_df = pd.merge(kmeans_results, metadata[['Sample Name', 'Factor Value[Spaceflight]']],
                           on='Sample Name', how='left')

    print("KMeans Results:")
    print(merged_df)
    print("\nFeature Importance (Approximate):")
    print(feature_importance_series)


    merged_df.to_csv('kmeans_results.csv', index=False)
    cluster_centers.to_csv('cluster_centers.csv', index=False)
    feature_importance_series.to_csv('feature_importance.csv', header=True)
    return merged_df, cluster_centers, feature_importance_series


if __name__ == "__main__":
    metadata_file = 'OSD-753Metadata.csv'
    data_file = 'LSDS-113_Immunoassay_Roy_ECLIA_Hippocampus_TRANSFORMED.csv'
    num_clusters = 2
    results_df, cluster_centers_df, feature_importance_series = analyze_data(metadata_file, data_file, num_clusters)
    
   
```
